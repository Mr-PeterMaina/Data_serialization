```python

# This  function  prints the user menu for my to do list.
def show_menu():
    print("HELLO, Welcome to a Cooking to-do list \n")
    print("1. add task")
    print("2. task completion")
    print("3. List tasks")
    print("4. delete task")
    print("5. Quit")
    print("\n")

# This functions prompts the user to input a task
def add_tasks():
    task = input("Enter the cooking task: ")
    tasks.append({"description":task, "completed" : False} )
    print("Task added successfully")

# This function marks the completed task
def mark_completed():
  task_no = int(input("Enter the number of task to mark complete:"))
  if task_no in range(len(tasks)):
    if not tasks[task_no]["completed"]:
            tasks[task_no]["completed"] = True
            print("Task to mark as completed!")
    else:
            print("This task is already marked as completed.")
  else:
      print("This is an invalid number")

# A function that lists all tasks that we have input
def list_tasks():
    if not tasks:
        print("No tasks in the list.")
    else:
        print("\t TASK LIST \t")
        for index, task in enumerate(tasks):
            status = "Completed" if task["completed"] else "Not Completed"
            print(f"{index}. {task['description']} - {status}")
            print("-----------------------------")

def remove_task():
    task_no = int(input("Enter the number of the task to delete: "))
    if task_no in range(len(tasks)):
        remove_task = tasks.pop(task_no)["description"]
        print(f"Task '{remove_task}' has been removed!")
    else:
        print("Invalid task number.")

# Function to handle action depending user input
def perform_tasks(choice):
    if choice == "1":
        add_tasks()
    elif choice == "2":
        mark_completed()
    elif choice == "3":
        list_tasks()
    elif choice == "4":
        remove_task()
    elif choice == "5":
        print("Goodbye!")
        quit()
    else:
        print("Invalid Input. Please try again.")

tasks = []

print("Welcome to my Cooking To-Do List!")

while True:
    show_menu()
    choice = input("Enter your choice (1-5): ")
    perform_tasks(choice)

```
